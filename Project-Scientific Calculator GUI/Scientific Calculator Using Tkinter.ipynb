# Ahm | Advanced Calculator

```python
import tkinter as tk
import math
def add_value(value):
    entry.insert(tk.END, value)

def backspace():
    text = entry.get()
    entry.delete(0, tk.END)
    entry.insert(tk.END, text[:-1])

def calculate():
    try:
        result = eval(entry.get())
        entry.delete(0, tk.END)
        entry.insert(tk.END, result)
    except:
        entry.delete(0, tk.END)
        entry.insert(tk.END, "Error")

def sin():
    value = float(entry.get())
    entry.delete(0, tk.END)
    entry.insert(tk.END, math.sin(math.radians(value)))

def cos():
    value = float(entry.get())
    entry.delete(0, tk.END)
    entry.insert(tk.END, math.cos(math.radians(value)))

def tan():
    value = float(entry.get())
    entry.delete(0, tk.END)
    entry.insert(tk.END, math.tan(math.radians(value)))

def sqrt():
    value = float(entry.get())
    entry.delete(0, tk.END)
    entry.insert(tk.END, math.sqrt(value))
    
def clear():
    entry.delete(0, tk.END)
window = tk.Tk()
window.title("AI | Advanced Calculator")
entry = tk.Entry(window, font=("Times New Roman", 20), justify="right", bd=10)
entry.grid(row=0, column=0, columnspan=4, padx=10, pady=20, ipady=15)
btn_style = {"font": ("Arial", 14), "bg": "#E4E7E7", "fg": "black", "padx": 20, "pady": 20}
buttons = [
    ("7", 1, 0), ("8", 1, 1), ("9", 1, 2), ("/", 1, 3),
    ("4", 2, 0), ("5", 2, 1), ("6", 2, 2), ("*", 2, 3),
    ("1", 3, 0), ("2", 3, 1), ("3", 3, 2), ("-", 3, 3),
    ("0", 4, 0), (".", 4, 1), ("=", 4, 2), ("+", 4, 3),
    ("(", 5, 0), (")", 5, 1), ("C", 5, 2), ("⌫", 5, 3),
    ("sin", 6, 0), ("cos", 6, 1), ("tan", 6, 2), ("√", 6, 3)
]
for text, row, col in buttons:
    if text == "=":
        cmd = calculate
    elif text == "C":
        cmd = clear
    elif text == "⌫":
        cmd = backspace
    elif text == "sin ()":
        cmd = sin
    elif text == "cos ()":
        cmd = cos
    elif text == "tan":
        cmd = tan
    elif text == "√sqrt":
        cmd = sqrt
    else:
        cmd = lambda x=text: add_value(x)
    tk.Button(window, text=text, command=cmd, **btn_style).grid(row=row, column=col, sticky="nsew")
    
window.mainloop()
```
